```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../Data/cleaned_data.csv')
df.head()
```

```python
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Property Prices')
plt.show()
```

```python
# Area vs Price
sns.scatterplot(data=df, x='area', y='price')
plt.title('Area vs Property Price')
plt.show()
```

```python
#Average Price by BHK
bhk_price = df.groupby('bhk')['price'].mean().reset_index()

sns.barplot(data=bhk_price, x='bhk', y='price')
plt.title('Average Price by BHK')
plt.show()
```

```python
df['bhk'].unique()
```

```python
#Price Distribution by Property Type
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='type', y='price')
plt.xticks(rotation=45)
plt.title('Price Distribution by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.show()
```

```python
#Property Status vs Price
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='status', y='price')
plt.xticks(rotation=30)
plt.title('Price Distribution by Property Status')
plt.xlabel('Status')
plt.ylabel('Price')
plt.show()
```

```python
# Correlation Heatmap (Numeric Features)
plt.figure(figsize=(8,6))
sns.heatmap(
    df[['bhk', 'area', 'age', 'price']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```
