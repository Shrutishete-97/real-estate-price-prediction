```python
import pandas as pd

df = pd.read_csv("../Data/Mumbai House Prices.csv")
df.head()
```

```python

print("Missing values:")
df.isnull().sum()

```

```python
print("\nData types:")
df.dtypes
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
print("Shape after removing duplicates:", df.shape)
```

```python
df.isnull().sum()
```

```python
df['price_unit'].unique()
```

```python
# Convert Crore to Lakh
df.loc[df['price_unit'] == 'Crore', 'price'] = df['price'] * 100

```

```python
print("Conversion complete!")
print(df['price_unit'].unique())
```

```python
# Optional: keep only one unit
df['price_unit'] = 'Lakh'
```

```python
df[['price', 'price_unit']].head()
```

```python
df['age'].unique()[:40]
```

```python
age_mapping = {
    'New': 0,
    'Resale': 1,
    'Unknown': 1
}

df['age'] = df['age'].map(age_mapping)
```

```python
df['age'].unique()
df['age'].dtype

```

```python
df.dtypes
```

```python
df = df[['bhk', 'area', 'price', 'age', 'type', 'locality', 'region', 'status']]

```

```python
df.head()
df.shape
```

```python
df.to_csv("../Data/cleaned_data.csv", index=False)
```
